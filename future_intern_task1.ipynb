```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('Iris.csv')
print(df.head())
```

```python
print("\nMissing Values:\n",df.isnull().sum())
```

```python
plt.figure(figsize=(10,6))
plt.hist(df['PetalLengthCm'], bins=20, color='skyblue', edgecolor='black')
plt.title('Distribution of Petal Length')
plt.xlabel('Petal Length (cm)')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

```python
species_count = df['Species'].value_counts()
plt.figure(figsize=(7, 5))
species_count.plot(kind='bar', color='lightgreen', edgecolor='black')

plt.title('Count of Each Iris Species')
plt.xlabel('Species')
plt.ylabel('Number of Samples')
plt.xticks(rotation=0)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
avg_petal_length = df.groupby('Species')['PetalLengthCm'].mean()

# Plot the bar chart
plt.figure(figsize=(7, 5))
avg_petal_length.plot(kind='bar', color='orange', edgecolor='black')

plt.title('Average Petal Length per Iris Species')
plt.xlabel('Species')
plt.ylabel('Average Petal Length (cm)')
plt.xticks(rotation=0)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```
